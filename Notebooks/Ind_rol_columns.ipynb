```python
import pickle
import pandas as pd
```

```python
def merge_expanding_means_ind(df):
    exp_lists = {}
    exp_ind = {}
    for column in rol_col_ind:
        exp_lists[column] = []
        exp_ind[column] = []

    exp_series_dict = {}
    for column in rol_col_ind:
        exp_series_dict[f'Season{column}'] = []
        
    for column in rol_col_ind:
    
        for team in df.groupby('PLAYER_ID').groups:
            team_rol = df.groupby('PLAYER_ID')[column].expanding().mean()[team].shift()
            exp_lists[column].extend(list(team_rol.values))
            exp_ind[column].extend(list(team_rol.index))
    
        exp_series_dict[f'Season{column}'] = pd.Series(exp_lists[column], exp_ind[column], name = f"{column} Season")
    
    for series in exp_series_dict:
        df = df.merge(exp_series_dict[series], left_index=True, right_index=True)
        
    return df
```

```python
rol_col_ind = ['OFF_RATING', 'DEF_RATING', 'NET_RATING'] 
```

```python
with open('data/pickles/starters21.p', 'rb') as readfile: 
    starters21 = pickle.load(readfile)
    
with open('data/pickles/starters20.p', 'rb') as readfile: 
    starters20 = pickle.load(readfile)
    
with open('data/pickles/starters19.p', 'rb') as readfile: 
    starters19 = pickle.load(readfile)
    
with open('data/pickles/starters18.p', 'rb') as readfile: 
    starters18 = pickle.load(readfile)
    
with open('data/pickles/starters17.p', 'rb') as readfile: 
    starters17 = pickle.load(readfile)
```

```python
starters20
```

```python
starters21.reset_index(inplace=True)
starters20.reset_index(inplace=True)
```

```python
starters21 = merge_expanding_means_ind(starters21)   
starters20 = merge_expanding_means_ind(starters20)
```

```python
starters19.reset_index(inplace=True)
starters18.reset_index(inplace=True)
starters17.reset_index(inplace=True)
```

```python
starters19 = merge_expanding_means_ind(starters19)
starters18 = merge_expanding_means_ind(starters18)
starters17 = merge_expanding_means_ind(starters17)
```
