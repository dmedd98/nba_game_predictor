The goal for this model for the model predicting the margin of victory is to be accurate enough where it could turn a profit if used to bet NBA games against the spread.  It is standard for a sportsbook to have -110 odds, meaning you need to risk $110 to win $100. This means if you win exactly half of your bets, you are going to lose money. With that in mind, you need to win 52.4% of the time to "break-even", this is illustrated below, with Expected Value (the measure of what a bettor can expect to win or lose per bet placed on the same odds time and time again) on the y-axis and Percentage of bets won on the x-axis

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.set_theme(style='darkgrid')

x = np.linspace(0, 1, 500)
x = np.ascontiguousarray(x, dtype=np.float32)
EV = ((x * 90.91) - ((1 - x) * 100))
break_even = 0 * x


plt.plot(x, EV, linewidth=3)
plt.plot(break_even, linewidth=3)
plt.scatter(.5238, 0, color = 'black', linewidths=15)
ax.annotate('Break Even Point (52.4%)', fontsize=15, xy=(.53, 0), xytext=(.65, 5),arrowprops = {'facecolor':'black'})


plt.xlabel('Percentage of Bets Won', fontsize=15)
plt.ylabel('Expected Value Per $100 Bet', fontsize=15)
plt.xlim(0, 1);
```

# Feature Engineering

This section will be an overview of how we used the dataframes containing box score data from over 3000 games to create feature columns for our model

```python
#Adding a PTS_ALLOWED columns to our Dataframes
splitdf21['PTS_ALLOWED'] = splitdf21['PTS'] - splitdf21['PLUS_MINUS']
splitdf20['PTS_ALLOWED'] = splitdf20['PTS'] - splitdf20['PLUS_MINUS']
splitdf19['PTS_ALLOWED'] = splitdf19['PTS'] - splitdf19['PLUS_MINUS']
splitdf18['PTS_ALLOWED'] = splitdf18['PTS'] - splitdf18['PLUS_MINUS']
splitdf17['PTS_ALLOWED'] = splitdf17['PTS'] - splitdf17['PLUS_MINUS']
```

Our model is only allowed to use information that would be known prior to the start of the game...for this reason, we are going to create new columns that have the average of each stats for the given teams' previous 7 games (using the last7games function) as well as columns containing the season average (season_stats function) for that statistic up to that point. 

```python
## List of the columns we want the rolling averages for
rol_columns = ['PTS', 'REB', 'AST','STL','BLK', 'TOV', 'PLUS_MINUS', 'FGA', 'FG_PCT', 'FG3A', 'FG3_PCT', 'FTA','FT_PCT', 'EFG_PCT_y', 'FTA_RATE', 'TM_TOV_PCT_y',
       'OREB_PCT_y', 'OPP_EFG_PCT', 'OPP_FTA_RATE', 'OPP_TOV_PCT',
       'OPP_OREB_PCT', 'REB_PCT', 'PIE', 'OFF_RATING', 'DEF_RATING', 'NET_RATING',
       'TS_PCT', 'PACE', 'AST_TOV', 'PTS_ALLOWED']
```

```python
def last7games(df):
    
    rol_lists_7 = {}
    rol_ind_7 = {}
    for column in rol_columns:
        rol_lists_7[column] = []
        rol_ind_7[column] = []
    
    series_dict_7 = {}
    for column in rol_columns:
        series_dict_7[f'Last_7{column}'] = []
        
    for column in rol_columns:
    
        for team in df.groupby('TEAM_ID').groups:
            team_rol = df.groupby('TEAM_ID')[column].rolling(7).mean()[team].shift()
            rol_lists_7[column].extend(list(team_rol.values))
            rol_ind_7[column].extend(list(team_rol.index))
    
        series_dict_7[f'Last_7{column}'] = pd.Series(rol_lists_7[column], rol_ind_7[column], name = f"{column} LAST 7")
    
    for series in series_dict_7:
        df = df.merge(series_dict_7[series], left_index=True, right_index=True)
        
    return df


def season_stats(df):
    exp_lists = {}
    exp_ind = {}
    for column in rol_columns:
        exp_lists[column] = []
        exp_ind[column] = []

    exp_series_dict = {}
    for column in rol_columns:
        exp_series_dict[f'Season{column}'] = []
        
    for column in rol_columns:
    
        for team in df.groupby('TEAM_ID').groups:
            team_rol = df.groupby('TEAM_ID')[column].expanding().mean()[team].shift()
            exp_lists[column].extend(list(team_rol.values))
            exp_ind[column].extend(list(team_rol.index))
    
        exp_series_dict[f'Season{column}'] = pd.Series(exp_lists[column], exp_ind[column], name = f"{column} Season")
    
    for series in exp_series_dict:
        df = df.merge(exp_series_dict[series], left_index=True, right_index=True)
        
    return df
```

```python
model_df21 = last7games(splitdf21)
model_df20 = last7games(splitdf20)
model_df19 = last7games(splitdf19)
model_df18 = last7games(splitdf18)
model_df17 = last7games(splitdf17)
```

```python
model_df21 = season_stats(model_df21)
model_df20 = season_stats(model_df20)
model_df19 = season_stats(model_df19)
model_df18 = season_stats(model_df18)
model_df17 = season_stats(model_df17)
```

We will use the functions below to create a column to note the home team, the team's winning percentage up to that point, and whether or not the team played the night before (back to back).

```python
## Function to create column that notes whether or not team is at home
import numpy as np

def note_home_team(df):
    df['HOME_TEAM'] = np.where(df['MATCHUP'].str.contains(' vs. '), 1, 0)

#Function will create columns noting whether or not the team won or lost
def track_win_loss(df):
    df['WIN'] = np.where(df['WL'].str.contains('W'), 1, 0)
    df['LOSS'] = np.where(df['WL'].str.contains('L'), 1, 0)

#Function will keep track of wins for that team up to that point
def keep_wins(df):
    wins_total = []
    wins_ind = []
    
    for team in df.groupby('TEAM_ID').groups:
        teamwins = df.groupby('TEAM_ID')['WIN'].expanding().sum()[team].shift()
        wins_total.extend(list(teamwins.values))
        wins_ind.extend(list(teamwins.index))
        
    wins_series = pd.Series(wins_total, wins_ind, name='WINS_SEASON')
    df['WINS_SEASON'] = wins_series
    

#Function will keep track of losses for that team up to that point
def keep_ls(df):
    loss_total = []
    loss_ind = []
    
    for team in df.groupby('TEAM_ID').groups:
        teamloss = df.groupby('TEAM_ID')['LOSS'].expanding().sum()[team].shift()
        loss_total.extend(list(teamloss.values))
        loss_ind.extend(list(teamloss.index))
        
    loss_series = pd.Series(loss_total, loss_ind, name='LOSS_SEASON')
    df['LOSS_SEASON'] = loss_series
    return (df)

#Function will keep track of the team's winning percentage going into the given game
def win_per(df):
    df['WIN_%'] = (df['WINS_SEASON'] / (df['LOSS_SEASON'] + df['WINS_SEASON'])) 
    


#Function to record date from the day before a given game and then a function to apply it to a new column
import datetime 
def yesterday(date):
    today = datetime.datetime.strptime(date, '%Y-%m-%d')
    yesterday = datetime.datetime.strftime(today - datetime.timedelta(1), '%Y-%m-%d')
    return yesterday

def yesterday_apply(df):
    df['previous_day'] = df.apply(lambda row: yesterday(date=row['GAME_DATE']), axis=1)

#Function to note whether or not the team played the on the day prior and then a function to apply it to a new column
def played_yesterday(df, date, team_id):
    filtered_df = df.loc[df.GAME_DATE == date]
    filtered_df = filtered_df.loc[filtered_df.TEAM_ID == team_id]
    if filtered_df.empty:
        return 0
    else:
        return 1
    
def played_yesterday_apply(df):
    df['back_to_back'] = df.apply(lambda row: played_yesterday(df, date=row['previous_day'], team_id = row['TEAM_ID']), axis=1)
```

Now that the team statistics are done, lets turn to the starters dataframe to create some rolling individual player statistics

```python
##This function works similiarly to the season_stats function from before, but it groups by player_id rather than team_id
rol_col_ind = ['OFF_RATING', 'DEF_RATING', 'NET_RATING', 'EFG_PCT',
 'TS_PCT', 'PIE', 'AST_PCT', 'REB_PCT'] 

def season_stats_player(df):
    exp_lists = {}
    exp_ind = {}
    for column in rol_col_ind:
        exp_lists[column] = []
        exp_ind[column] = []

    exp_series_dict = {}
    for column in rol_col_ind:
        exp_series_dict[f'Season{column}'] = []
        
    for column in rol_col_ind:
    
        for team in df.groupby('PLAYER_ID').groups:
            team_rol = df.groupby('PLAYER_ID')[column].expanding().mean()[team].shift()
            exp_lists[column].extend(list(team_rol.values))
            exp_ind[column].extend(list(team_rol.index))
    
        exp_series_dict[f'Season{column}'] = pd.Series(exp_lists[column], exp_ind[column], name = f"{column} Season")
    
    for series in exp_series_dict:
        df = df.merge(exp_series_dict[series], left_index=True, right_index=True)
        
    return df
```

Now we will add the predictions the model has made for each of the game's in the test set, and then add the monetary result and rolling profit/loss we would have if we had bet the model's recomendation for each game in the test set

```python
#Model prediction column
test_set['Test_predictions'] = lrgs.predict(X_sc_test)
```

```python
test_set.reset_index(inplace=True)
#Create a column that displays money won/lost for given game
test_set['Monetary Result'] = 0
for ind in test_set.index:
    # Prediction is correct
    if test_set['Test_predictions'].iloc[ind] == test_set['WIN_Home'].iloc[ind]:
        #HomeTeam Won
        if test_set['WIN_Home'].iloc[ind] == 1:
            #HomeTeam Favored
            if test_set['ML_Home'].iloc[ind] < 0:
                test_set.at[ind, 'Monetary Result'] = (100 / ((abs(test_set['ML_Home'].iloc[ind]))/100))
            #HomeTeam Underdog
            else:
                test_set.at[ind, 'Monetary Result'] = (100 * ((test_set['ML_Home'].iloc[ind])/100))
        #AwayTeam Won
        elif test_set['WIN_Home'].iloc[ind] == 0:
            #AwayTeam Favored
            if test_set['ML_Away'].iloc[ind] < 0:
                test_set.at[ind, 'Monetary Result'] = (100 / ((abs(test_set['ML_Away'].iloc[ind]))/100))
            #AwayTeam Underdog
            else:
                test_set.at[ind, 'Monetary Result'] = (100 * ((test_set['ML_Away'].iloc[ind])/100))
    #Prediction is wrong
    else:    
        test_set.at[ind, 'Monetary Result'] = -100
            
```

```python
test_set['Profit'] = test_set['Monetary Result'].cumsum()
```

```python
#Checking the accuracy of the model
accuracy = len(test_set[test_set['Test_predictions'] == test_set['WIN_Home']])/len(test_set)
print(f'Accuracy of model picking the winning team: {accuracy}')
```
