```python
even = np.random.choice([0, 1], size=100000, p=[.5, .5])
fiftyfive = np.random.choice([0, 1], size=100000, p=[.45, .55])
fiftythree = np.random.choice([0, 1], size=100000, p=[.47, .53])
```

```python
profit = 0
profit_track = []
listn = []
for n in range(0, len(even)):
    listn.append(n)
    bet = ((even[n] * 90.91) - ((1 - even[n]) * 100))
    profit += bet
    profit_track.append(profit)

profitdf = pd.DataFrame(profit_track, listn, columns = ['profit'])
```

```python
ev = []
for n in range(0, len(even)):
    ev1 = (profitdf['profit'].iloc[n] / n)
    ev.append(ev1)

profitdf['EV'] = ev
```
